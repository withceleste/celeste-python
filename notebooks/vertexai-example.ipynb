### In diesem Notizbuch wird die grundlegende Verwendung von VertexAI mit Celeste demonstriert.
1. Set correct vertexai project number including location and flag
2. Turn on the web_search (online rag)

```python
from celeste import create_client, Modality, Operation, Provider
from celeste.providers.google.auth import GoogleADC
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

GOOGLE_CLOUD_PROJECT = os.getenv("GOOGLE_CLOUD_PROJECT", "xxxx")
GOOGLE_CLOUD_LOCATION = os.getenv("GOOGLE_CLOUD_LOCATION", "global")
GOOGLE_GENAI_USE_VERTEXAI="True"
```

```python
client = create_client(
    modality=Modality.TEXT,
    operation=Operation.GENERATE,
    provider=Provider.GOOGLE,
    model="gemini-3.1-pro-preview",
        auth=GoogleADC(
        project_id=GOOGLE_CLOUD_PROJECT, 
        location=GOOGLE_CLOUD_LOCATION             
    )
)
```

```python
response = await client.generate(
    "Um Quantencomputing zu erklären, könnt ihr mir Zitatlinks anbieten, wenn ihr fertig seid.",
    web_search=True,
)
print(response.content)
```
